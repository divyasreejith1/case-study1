# Perform meaningful visualizations using the dataset. Bring at least 3 visualisations.

```python
#seaborn lineplot of sepal length and petal length of iris
```

```python
sns.lineplot(x=data['SL'],y=data['PL'],hue=data['Classification'])
plt.title('sepal length vs petal length of iris')
plt.show()
```

```python
#seaborn boxplot of classification and petal width of iris
```

```python
sns.boxplot(x=data['Classification'],y=data['PW'])
plt.title('Classification of iris vs petal width')
plt.show()
```

```python
#matplotlib scatterplot of sepal width and petal width
```

```python
plt.scatter(data['SW'],data['PW'])
plt.title('sepal width vs petal width of iris')
plt.xlabel('SW')
plt.ylabel('PW')
plt.show()
```

```python
#Seaborn violinplot of classification of iris and petal length
```

```python
sns.violinplot(x=data['Classification'],y=data['PL'])
plt.title('classification of iris vs petal length')
plt.show()
```
